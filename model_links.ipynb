```python
from transformers import AutoModel
```

```python
total_params = 0
attention_params = 0
kvq_params = 0
embed_params = 0
model = AutoModel.from_pretrained("facebook/opt-13b") 
for name, parameter in model.named_parameters():
    # Count the total number of parameters
    param_count = parameter.numel()
    total_params += param_count
    # Check if the layer is an attention layer
    if "embed" in name:
        embed_params += param_count
    if "k_proj" in name or "v_proj" in name or "q_proj" in name:
        kvq_params += param_count
        print(f"{name}: {param_count:,} parameters")
    elif "attention" in name or "attn" in name:
        attention_params += param_count
        print(f"{name}: {param_count:,} parameters")
    else:
        print(f"{name}: {param_count:,} parameters")
```

```python
print(total_params)
print(embed_params, embed_params/total_params)
print(attention_params, attention_params/total_params)
print(kvq_params, kvq_params/total_params)
```

```python
from scipy import spatial
from scipy.special import softmax
import random
import numpy
```

```python
test = 1000
cosines = []
euclideans = []
minkowski = []
for i in range(test):
    list1 = softmax([random.random() for _ in range(1000)])
    list2 = softmax([random.random() for _ in range(1000)])
    result = 1 - spatial.distance.cosine(list1, list2)
    result_euc = 1 - spatial.distance.euclidean(list1, list2)
    result_min = 1 - spatial.distance.minkowski(list1, list2, p=2)
    cosines.append(result)
    euclideans.append(result_euc)
    minkowski.append(result_min)

print("cosine results")    
print(sum(cosines)/test)
print(numpy.median(cosines))
print(min(cosines))
print(max(cosines))
print("\n")
print("euclidean results")
print(sum(euclideans)/test)
print(numpy.median(euclideans))
print(min(euclideans))
print(max(euclideans))
print("\n")
print("minkowski results")
print(sum(minkowski)/test)
print(numpy.median(minkowski))
print(min(minkowski))
print(max(minkowski))
```
